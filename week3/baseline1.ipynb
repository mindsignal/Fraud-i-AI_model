### trans_date_trans_time

```python
train['datetime'] = pd.to_datetime(train['trans_date_trans_time'])
```

```python
# 시(hour) 추출 (0~23)
train['hour'] = train['datetime'].dt.hour

# 요일 추출: Monday=0, Sunday=6
train['day_of_week'] = train['datetime'].dt.dayofweek

# 월 추출: 1월=1, 12월=12
train['month'] = train['datetime'].dt.month
```

#### 1. sin/cos 사용

```python
train['trans_hour_sin'] = np.sin(2 * np.pi * train['hour'] / 24)
train['trans_hour_cos'] = np.cos(2 * np.pi * train['hour'] / 24)
```

#### 2. is_night 사용

```python
train['is_night'] = train['hour'].isin([22, 23, 0, 1, 2, 3]).astype(int)
```

#### 3. is_high_risk_period 사용

```python
# 금(4), 목(3), 수(2) + 심야 시간대(0~4, 21~23)
high_risk_days = [2, 3, 4]  # 수, 목, 금
high_risk_hours = list(range(0, 5)) + [21, 22, 23]

train['is_high_risk_period'] = (
    train['day_of_week'].isin(high_risk_days) & train['hour'].isin(high_risk_hours)
).astype(int)
```

### cc_num

```python
from bisect import bisect_left, bisect_right

def count_past_days_fast(df, days):
    results = []
    grouped = df.groupby('cc_num')

    for card, group in grouped:
        times = group['trans_date_trans_time'].tolist()
        counts = []
        for i, t in enumerate(times):
            start_time = t - pd.Timedelta(days=days)
            # 왼쪽 경계 (start_time 이상)
            left_idx = bisect_left(times, start_time)
            # 오른쪽 경계는 현재 거래 바로 전(i번째 거래 제외)
            right_idx = i
            counts.append(right_idx - left_idx)
        results.extend(counts)
    return results
```

```python
train['trans_date_trans_time'] = pd.to_datetime(train['trans_date_trans_time'])
train = train.sort_values(['cc_num', 'trans_date_trans_time']).reset_index(drop=True)

train['cnt_1d'] = count_past_days_fast(train, 1)
train['cnt_7d'] = count_past_days_fast(train, 7)
train['cnt_30d'] = count_past_days_fast(train, 30)

train['next_trans_time'] = train.groupby('cc_num')['trans_date_trans_time'].shift(-1)
train['time_since_last_trans'] = (train['trans_date_trans_time'] - train.groupby('cc_num')['trans_date_trans_time'].shift(1)).dt.total_seconds()

# 또는 마지막 거래 이후 경과 시간(다음 거래까지 시간 간격)
train['time_until_next_trans'] = (train['next_trans_time'] - train['trans_date_trans_time']).dt.total_seconds()

# 결측치 처리 (time_since_last_trans 첫 거래는 NaN, 0으로 채우거나 다른 방식)
train[['time_since_last_trans', 'time_until_next_trans']] = train[['time_since_last_trans', 'time_until_next_trans']].fillna(0)
```

### dob

#### 1. age 그대로 사용

```python
train['dob'] = pd.to_datetime(train['dob'])

from dateutil.relativedelta import relativedelta

def calculate_age(born, trans_date):
    return relativedelta(trans_date, born).years

train['age'] = train.apply(lambda x: calculate_age(x['dob'], x['trans_date_trans_time']), axis=1)
```

#### 2. group으로 나눌 시

```python
train['age_group'] = (train['age'] // 10) * 10
```

```python
# 1. 스무딩 인코딩 함수 정의
def target_smooth_encoding(df, column, target, smoothing):
    global_mean = df[target].mean()
    agg = df.groupby(column)[target].agg(['mean', 'count'])
    agg['smoothed'] = (
        (agg['mean'] * agg['count'] + global_mean * smoothing) /
        (agg['count'] + smoothing)
    )
    return df[column].map(agg['smoothed']).fillna(global_mean)

# 2. 'job'과 'city_state'에 대해 스무딩 인코딩 적용
df['job_enc'] = target_smooth_encoding(df, 'job', 'is_fraud', smoothing=100)
df['city_state_enc'] = target_smooth_encoding(df, 'city_state', 'is_fraud', smoothing=100)
df.head()
```

# 증강 결과 비교

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

# 1. 평가 함수 정의
def evaluate_model(y_true, y_pred, y_proba):
    print("🔍 Classification Report:")
    print(classification_report(y_true, y_pred, digits=4))

    roc = roc_auc_score(y_true, y_proba)
    print(f"\n✅ ROC AUC: {roc}")

    cm = confusion_matrix(y_true, y_pred)
    print("📉 Confusion Matrix:")
    print(cm)

# 2. Label Encoding (category, gender)
for col in ['category', 'gender']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# 3. 사용하지 않을 컬럼 제거 (문자형 원본들)
drop_cols = ['job', 'city_state']
X = df.drop(columns=['is_fraud'] + drop_cols)
y = df['is_fraud']

# 4. 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 5. LightGBM 학습
model = LGBMClassifier(random_state=42)
model.fit(X_train, y_train)

# 6. 예측 및 평가
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba)
```

```python
from sdv.single_table.ctgan import CTGAN
# 1. 평가 함수
def evaluate_model(y_true, y_pred, y_proba):
    print("🔍 Classification Report:")
    print(classification_report(y_true, y_pred, digits=4))

    roc = roc_auc_score(y_true, y_proba)
    print(f"\n✅ ROC AUC: {roc}")

    cm = confusion_matrix(y_true, y_pred)
    print("📉 Confusion Matrix:")
    print(cm)

# 2. category, gender 인코딩 (LabelEncoder → fit 저장)
le_category = LabelEncoder()
le_gender = LabelEncoder()

df['category'] = le_category.fit_transform(df['category'].astype(str))
df['gender'] = le_gender.fit_transform(df['gender'].astype(str))

# 3. 사기 거래만 추출해서 CTGAN 학습
fraud_df = df[df['is_fraud'] == 1].copy()

# 문자열로 변환 (CTGAN 요구 사항)
fraud_df['category'] = fraud_df['category'].astype(str)
fraud_df['gender'] = fraud_df['gender'].astype(str)
fraud_df = fraud_df.drop(columns=['job', 'city_state'])

# CTGAN 학습
model = CTGAN(epochs=100)
model.fit(fraud_df, discrete_columns=['category', 'gender'])

# 4. 샘플 생성 (예: 6000건)
synthetic_fraud = model.sample(n=6000)
synthetic_fraud['is_fraud'] = 1

# 5. LabelEncoder로 다시 변환 (기존 df와 동일하게 맞추기)
synthetic_fraud['category'] = le_category.transform(synthetic_fraud['category'])
synthetic_fraud['gender'] = le_gender.transform(synthetic_fraud['gender'])

# 6. 기존 데이터 + 증강 데이터 병합
augmented_df = pd.concat([df, synthetic_fraud], ignore_index=True)
augmented_df = augmented_df.sample(frac=1, random_state=42).reset_index(drop=True)

# 7. 학습에 사용할 컬럼 정의 (문자형 제거)
drop_cols = ['is_fraud', 'job', 'city_state']  # 이미 인코딩된 버전만 사용
X = augmented_df.drop(columns=drop_cols)
y = augmented_df['is_fraud']

# 8. train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 9. LightGBM 학습
model = LGBMClassifier(random_state=42)
model.fit(X_train, y_train)

# 10. 평가
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

evaluate_model(y_test, y_pred, y_proba)
```
